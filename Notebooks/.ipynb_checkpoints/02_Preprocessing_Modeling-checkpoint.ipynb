## Ce notebook a pour objectifs  
- Nettoyage des tweets
- Suppression des chiffres, ponctuation et mots courts
- Suppression des stopwords
- Stemming et reconstitution du texte
- Visualisation des tokens

```python
import pandas as pd
import numpy as np
import re
import string
import nltk

nltk.download('stopwords')
nltk.download('wordnet')

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def basic_tokenizer(text):
    return re.findall(r'\b\w+\b', text.lower())
```

```python
df = pd.read_csv("../data/train.csv")
df = df[["text", "target"]]
```

### Nettoyage du texte 

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", '', text)
    text = re.sub(r'@\w+|#', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'\d+', '', text)
    text = ' '.join([word for word in text.split() if len(word) > 2])
    return text
```

### Application + Lemmatisation + Stopwords

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = clean_text(text)
    tokens = basic_tokenizer(text)  
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return " ".join(tokens)

df["clean_text"] = df["text"].apply(preprocess)
```

### Visualisation des tokens

```python
all_words = ' '.join(df["clean_text"])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_words)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("WordCloud des mots les plus fréquents")
plt.show()
```
